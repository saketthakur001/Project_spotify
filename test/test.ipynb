```python
import psycopg2
import main

def store_user_data_to_postgresql(friends_activity_json, 
                                  database_name='friends_activity',
                                  user="postgres",
                                  password="spotify",
                                  host="localhost",
                                  port="5432"):
    
    # Connect to the PostgreSQL server
    conn = psycopg2.connect(dbname=database_name, user=user, password=password, host=host, port=port)
    cur = conn.cursor()

    # Create tables in PostgreSQL
    
    # Users table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS users (
        user_id SERIAL PRIMARY KEY,
        user_url TEXT NOT NULL,
        user_name TEXT NOT NULL,
        user_image_url TEXT
    )
    ''')

    # Albums table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS albums(
        album_id SERIAL PRIMARY KEY,
        album_uri TEXT NOT NULL,
        album_name TEXT NOT NULL
    )
    ''')

    # Artists table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS artists(
        artist_id SERIAL PRIMARY KEY,
        artist_uri TEXT NOT NULL,
        artist_name TEXT NOT NULL
    )
    ''')

    # Tracks table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS tracks (
        track_id SERIAL PRIMARY KEY,
        track_uri TEXT NOT NULL,
        track_name TEXT NOT NULL,
        track_image_url TEXT NOT NULL,
        album_id INTEGER REFERENCES albums(album_id),
        artist_id INTEGER REFERENCES artists(artist_id)
    )
    ''')

    # Context table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS context(
        context_id SERIAL PRIMARY KEY,
        context_uri TEXT NOT NULL,
        context_name TEXT NOT NULL,
        context_index INTEGER NOT NULL
    )
    ''')

    # Streamings table
    cur.execute('''
    CREATE TABLE IF NOT EXISTS streamings(
        user_id INTEGER REFERENCES users(user_id),
        track_id INTEGER REFERENCES tracks(track_id),
        timestamp TIMESTAMP NOT NULL,
        context_id INTEGER REFERENCES context(context_id)
    )
    ''')

    # Commit the table creation
    conn.commit()

    # ... (Rest of the code for inserting data remains largely the same, but with minor adjustments)
    # For now, let's just handle the connection and table creation logic

    # Close the cursor and connection
    cur.close()
    conn.close()

# This function now sets up the PostgreSQL tables. Data insertion logic will follow.
store_user_data_to_postgresql(main.get_friends_activity_json())
```

```python
import os

# Path to the pgAdmin4 installation directory
pgadmin_path = "C:\\Program Files\\PostgreSQL\\15\\pgAdmin 4\\web"

# Files to search for
config_files = ["config_local.py", "config.py"]

# Flag to determine if a change was made
modified = False

for config_file in config_files:
    file_path = os.path.join(pgadmin_path, config_file)
    
    # Check if file exists
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            lines = f.readlines()
        
        # Look for the line with DEFAULT_SERVER_PORT and change its value
        for index, line in enumerate(lines):
            if "DEFAULT_SERVER_PORT" in line:
                print(f"Changing port in {config_file}...")
                lines[index] = "DEFAULT_SERVER_PORT = 5051\n"
                modified = True
                break
        
        # If a change was made, write the updated lines back to the file
        if modified:
            with open(file_path, 'w') as f:
                f.writelines(lines)
            print(f"Port changed in {config_file} to 5051.")
            break

if not modified:
    print("Could not find DEFAULT_SERVER_PORT in the configuration files.")
else:
    print("Please try launching pgAdmin4 again.")
```
